```python
# Step 1: Install Hugging Face libraries
!pip install transformers datasets

# Step 2: Import the necessary modules
from transformers import pipeline

# Step 3: Create a function that performs different NLP tasks based on user input
def nlp_task_selector(task_type, user_input):
    # Dictionary to map task type to corresponding Hugging Face pipeline
    task_pipelines = {
        "sentiment-analysis": pipeline("sentiment-analysis"),
        "summarization": pipeline("summarization"),
        "text-generation": pipeline("text-generation", model="gpt2")
    }

    # Get the appropriate pipeline for the task
    if task_type in task_pipelines:
        nlp_pipeline = task_pipelines[task_type]
        # Run the task
        result = nlp_pipeline(user_input)
        return result
    else:
        return f"Task '{task_type}' not supported. Please choose from 'sentiment-analysis', 'summarization', or 'text-generation'."

# Step 4: Get user input for the task and text
task_type = input("Enter the NLP task you want to perform (sentiment-analysis, summarization, text-generation): ").lower()
user_input = input("Enter the text you want to process: ")

# Step 5: Call the function and display the result
result = nlp_task_selector(task_type, user_input)
print("Result:", result)
```

```python
# Step 1: Install the Required Libraries
!pip install gradio transformers datasets

# Step 2: Import the necessary libraries
import gradio as gr
from transformers import pipeline

# Step 3: Define the NLP pipelines for different tasks
sentiment_pipeline = pipeline("sentiment-analysis")
summarization_pipeline = pipeline("summarization")
text_generation_pipeline = pipeline("text-generation", model="gpt2")

# Step 4: Define a function to handle user input and process the NLP task
def nlp_task(task, text):
    if task == "Sentiment Analysis":
        # Sentiment analysis: Returns the label (positive/negative) and confidence score
        return sentiment_pipeline(text)[0]
    elif task == "Summarization":
        # Summarization: Summarizes long text into a shorter form
        return summarization_pipeline(text)[0]['summary_text']
    elif task == "Text Generation":
        # Text Generation: Generate text based on the user's prompt
        return text_generation_pipeline(text, max_length=50)[0]['generated_text']
    else:
        return "Invalid Task Selected"

# Step 5: Create the Gradio interface
# Use gr.Dropdown instead of gr.inputs.Dropdown
interface = gr.Interface(
    fn=nlp_task,  # The function that processes the input
    inputs=[
        gr.Dropdown(choices=["Sentiment Analysis", "Summarization", "Text Generation"], label="NLP SEY KYA KAAM KARANA HAI SELECT KARO"),
        gr.Textbox(lines=5, placeholder=" YAHA APNA TEXT LIKHO", label="Input Text")
    ],
    outputs="text",  # The output will be displayed as text
title="WORDVERSE(😊)",
# Removed invalid 'PARAGRAPH' parameter, use 'description' instead
    description="Alok'S NLP",
    # description= "Choose a task (Sentiment Analysis, Summarization, or Text Generation) and input your text. The model will process it and display the result."
)
# Step 6: Launch the Gradio web app
interface.launch(share=True)  # The share=True option gives you a public link to the app
```
